### Identify the publisher that has released the greatest number of books with more than 50 pages.

```python
query = '''
select
publishers.publisher as biggest_publisher_above_50,
count(book_id) as book_cnt
from
books 
inner join publishers on  publishers.publisher_id = books.publisher_id
where
books.num_pages > 50
group by
publishers.publisher
order by
book_cnt desc


        '''
queryResult(query)
```

I will ad a count of small publishers to assess the distribution better. Meantime, it seems like there is a large number of publishers who are publishing a small number of books.

```python
query = '''
with new_books as
(select
publishers.publisher as biggest_publisher_above_50,
count(book_id) as book_cnt
from
books 
inner join publishers on  publishers.publisher_id = books.publisher_id
where
books.num_pages > 50
group by
publishers.publisher)
select
count(book_cnt) 
from 
new_books
where 
book_cnt <=5

        '''
queryResult(query)
```

We see that this distributiun has a very long tail - where a few tens of publishers publish more than 5 books and about 290 publish up to 5 books. Penguin has 2 subsidaries ranked in the top 5 (they have a close id umber which seems to hint that this is a relarted entity - TBD). ### Identify the author with the highest average book rating (look only at books with at least 50 ratings).

```python
query = '''
select
authors.author as highest_rank_author,
avg(ratings.rating) as mean_rating
from
books 
inner join authors on  authors.author_id = books.author_id
inner join ratings on ratings.book_id = books.book_id
group by
authors.author
having
count(rating) > 50
order by
mean_rating desc
limit 5;


        '''
queryResult(query)
```

The most popular author is J.K Rowling. Mary GrandPre is the illustrater of her books in the U.S. Editors, translators, ilustrators etc. are sometimes included in the writers value next to him. 

It is also worth noting that a Vietnamese writer is ranked 3rd by readers, and this could indicate a potentialy interesting market to target.

### Find the average number of text reviews among users who rated more than 50 books.


```python
query = '''
with rating_over_50 as
(select
username,
count(rating) as rating_cnt
from ratings
group by 
username
having 
count(rating) > 50)
select
count (reviews.text) / count(distinct (reviews.username))  as avg_review_num
from
reviews
inner join rating_over_50 on rating_over_50.username = reviews.username
where 
reviews.username in (select username from rating_over_50)



        '''
queryResult(query)
```

We can expect our heavy raters to be heavy reviewrs too - those who rated over 50 times wrote about half the time they rated - 24 reviews on average.
